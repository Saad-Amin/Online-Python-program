```python
# Question no: 7
marks_Eng = int(input("Enter marks of english: "))
marks_Isl = int(input("Enter marks of islamiat: "))
marks_phy = int(input("Enter marks of physics: "))
marks_com = int(input("Enter marks of computer: "))
marks_urdu = int(input("Enter marks of urdu: "))
total_marks = marks_Eng + marks_Isl + marks_phy+ marks_com + marks_urdu
percentage = (total_marks * 100)/500;
print("Percentage is ",percentage ,"%")

if percentage >= 80 and percentage <= 100:
    print("A Grade")
elif percentage >= 70 and percentage <= 80:
    print("B Grade")
elif percentage >= 60 and percentage <= 70:
    print("C Grade")
else:
    print("Fail")
```
